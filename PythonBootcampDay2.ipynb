# Guessing game (computer generate random number from 0 to 10, ask user the required number of trials, if the user enter 0, they the program generate random number of trials)

```python
import random
N = random.randrange(10)
trails = int(input("Enter numbers of trys : "))
if trails == 0 or trails >= 5:
  trails = random.randint(1,5)
print("you have" , trails ,"trys")
A = input
while A!= N and trails != 0:
  A = int(input("Guess the number 0 - 10 ? "))
  if A < N:
    trails-=1
    print("Low Try Again","you have" , trails ,"trys")
  if A > N:
    trails-=1
    print("High Try Agian", "you have" , trails ,"trys")
  
if A == N:
  print("***********Correct***********")  
if trails == 0:
  print("Your out of Trys Play again")
```
